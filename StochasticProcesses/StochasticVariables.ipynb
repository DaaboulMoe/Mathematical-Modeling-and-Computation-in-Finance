```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from mpl_toolkits import mplot3d
```

```python
mu = 10
sigma = 1
i = complex(0,1)

pdf = lambda x : norm.pdf(x, mu, sigma)
cdf = lambda x : norm.cdf(x, mu, sigma)
Chf = lambda u: np.exp(i * mu * u - sigma * sigma * u * u / 2.0)

x = np.linspace(5,15,100)
u = np.linspace(0,5,250)

#Plot Scaled Pdf and Cdf
plt.figure(figsize=(6,4))
plt.plot(x, pdf(x)/pdf(x).max(), label = "PDF (Normalized)" )
plt.plot(x, cdf(x), label = "CDF") 
plt.title("PDF and CDF")
plt.legend()
plt.show()  
```

```python
ChValues = Chf(u)
  
ax = plt.axes(projection='3d')
x = np.real(ChValues)
y = np.imag(ChValues)
ax.plot3D(u, x, y, 'blue')
ax.view_init(20, -120)
# Set the title
ax.set_title("Characteristic Function")

# Add labels
ax.set_xlabel('u')
ax.set_ylabel('Real')
ax.set_zlabel('Imaginary')

plt.show()
```
